```python
# Load in packages

## Load in scrapy for web scraping, and other relevant functions
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.spiders import CrawlSpider, Rule
from scrapy.linkextractors import LinkExtractor
from scrapy.selector import Selector

## Load in package for regex
import re

```

```python
#Define web scraper

class scraper(CrawlSpider):
    name = "boe_workingpapers"
    
    # Start web scraper off on a sitemap with links to all BoE working papers
    start_urls = [
           "https://www.bankofengland.co.uk/sitemap/staff-working-paper"
        ]

    # Use regex to define a rule, keeping only URLs which are related to working paper publications
    rules = (
            Rule(LinkExtractor(allow=(r'https:\/\/www.bankofengland.co.uk\/working-paper\/\d+\/(\w+|-)+',)), callback = 'parse', follow= True) ,
    )
    
    def parse(self, response):
        
        # Retrieve links in the main body of the summary
        pdf_extension = response.xpath('//div[@class="content-block"]//a//@href').extract()
        
        # Filter links to only those that are pdfs (excludes .zip files containing code for working papers, and contact links
        if re.search(r'(.+).pdf(.+)', pdf_extension)
            pdf_url = "https://www.bankofengland.co.uk" + pdf_extension 
        
        yield { 'pdf': pdf_url  
                        }
   
```

```python
process = CrawlerProcess()

process.crawl(scraper)
process.start()
```
